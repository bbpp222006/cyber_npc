```python
# Get TTS audio data asynchronously using httpx
import httpx


async def get_tts_audio(text: str):
    """
    异步调用TTS服务以获取音频数据并返回为bytes
    """
    url = "http://192.10.221.53:30004/run-inference"
    payload = {
        "target_text_content": text
    }

    headers = {
        "Content-Type": "application/json"
    }

    async with httpx.AsyncClient() as client:
        try:
            response = await client.post(url, json=payload,timeout=10)
            if response.status_code == 200:
                audio_data = response.content
                return {
                    "audio_data": audio_data,
                    "text": text
                }
            else:
                return None
        except Exception as e:
            return e
await get_tts_audio("哈哈哈")
```

```python
from pydub import AudioSegment
from IPython.display import Audio

# 输入文件和参数
input_mp3 = "fiish_speech/data/1.mp3"  # 输入文件路径
output_mp3 = "fiish_speech/data/1_1.mp3"  # 输出文件路径
db_increase = 12  # 增加的分贝值

try:
    # 加载 MP3 文件
    audio = AudioSegment.from_file(input_mp3, format="mp3")

    # 增加音量
    louder_audio = audio + db_increase

    # 导出处理后的文件
    louder_audio.export(output_mp3, format="mp3")
    print(f"音量已增加 {db_increase} 分贝，文件已保存到: {output_mp3}")

    # 预览音频
    display(Audio(output_mp3))
except Exception as e:
    print(f"处理音频文件时出错: {e}")
```

```python
from pydub import AudioSegment

audio = AudioSegment.from_file("fiish_speech/data/1_1.mp3", format="mp3")
    
# Calculate the duration to keep
duration_to_keep = len(audio) - 500  # Duration in milliseconds (1000 ms = 1 second)

if duration_to_keep <= 0:
    print("Error: The audio file is less than 1 second long.")

# Trim the audio
trimmed_audio = audio[:duration_to_keep]

# Export the modified audio
trimmed_audio.export("fiish_speech/data/1_2.mp3", format="mp3")
display(Audio("fiish_speech/data/1_2.mp3"))
```

```python
import librosa
import noisereduce as nr
import soundfile as sf  # librosa.output 已弃用，改用 soundfile 写文件

# 1. 读入音频
#  - sr=None 表示使用音频本身的采样率，不做重采样
#  - librosa 能读取 MP3，但需要 ffmpeg 支持
audio_path = "audio/1_1.mp3"
y, sr = librosa.load(audio_path, sr=None)

# 2. 进行降噪
reduced_noise = nr.reduce_noise(y=y, sr=sr)

# 3. 写出降噪后的音频 (保存为 WAV)
sf.write("audio/output_denoised.wav", reduced_noise, sr)
display(Audio("audio/output_denoised.wav"))
```
